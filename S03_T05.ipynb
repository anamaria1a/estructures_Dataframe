- Taula de les aerolínies amb més endarreriments acumulats



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
delayed_f = pd.read_csv("C:/Users/ANA MARIA/Downloads/DelayedFlights.csv") 


delayed_f["AccumulatedDelays"] = round(delayed_f["DepDelay"]+delayed_f["CarrierDelay"]+delayed_f["ArrDelay"]+delayed_f["WeatherDelay"]+delayed_f["SecurityDelay"]+delayed_f["NASDelay"]+delayed_f["LateAircraftDelay"])
delayed_f_group = delayed_f.groupby(["UniqueCarrier"]).agg({"AccumulatedDelays": np.median})
delayed_f_group.sort_values(by="AccumulatedDelays", ascending=False)[0:50]
```
